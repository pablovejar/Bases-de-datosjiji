```python
%%sql
DROP VIEW IF EXISTS comuna_mujeres;
CREATE VIEW comuna_mujeres AS
SELECT codigocomuna, (mujeres*1.0) / total AS "ratiomujeres"
FROM demografia
```

```python
%sql SELECT * FROM comuna_mujeres
```

```python
%%sql
DROP VIEW IF EXISTS comuna_mujeres_delincuencia;
CREATE VIEW comuna_mujeres_delincuencia AS
SELECT comuna_mujeres.codigocomuna, comuna_mujeres.ratiomujeres, casen.tasadelincuencia
FROM comuna_mujeres, casen
WHERE comuna_mujeres.codigocomuna = casen.codigocomuna
```

```python
%sql SELECT * FROM comuna_mujeres_delincuencia
```

```python
%%sql
SELECT
    CASE
        WHEN ratiomujeres > 0.5 THEN 'mujeres'
        ELSE 'hombres'
    END SexoMayoritario,
    COUNT(codigocomuna) AS 'Cantidad de comunas',
    AVG(tasadelincuencia) AS 'Tasa delincuencia promedio'
FROM comuna_mujeres_delincuencia
GROUP BY SexoMayoritario
```

```python
%%sql
SELECT
    CASE
        WHEN ratiomujeres < 0.4 THEN '0 - 0.4'
        WHEN ratiomujeres >= 0.4 AND ratiomujeres < 0.46 THEN '0.4 - 0.46'
        WHEN ratiomujeres >= 0.46 AND ratiomujeres < 0.48 THEN '0.46 - 0.48'
        WHEN ratiomujeres >= 0.48 AND ratiomujeres < 0.49 THEN '0.48 - 0.49'
        WHEN ratiomujeres >= 0.49 AND ratiomujeres < 0.5 THEN '0.45 - 0.5'
        WHEN ratiomujeres >= 0.5 AND ratiomujeres < 0.51 THEN '0.5 - 0.51'
        WHEN ratiomujeres >= 0.51 AND ratiomujeres < 0.52 THEN '0.51 - 0.52'
        ELSE '0.52 - 1'
    END ProporcionMujeres,
    COUNT(codigocomuna) AS 'Cantidad de comunas',
    AVG(tasadelincuencia) AS 'Tasa delincuencia promedio'
FROM comuna_mujeres_delincuencia
GROUP BY ProporcionMujeres
```
